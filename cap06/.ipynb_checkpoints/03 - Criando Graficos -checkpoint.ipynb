### Criando Gráficos atráves de dados do BD

```python
import sqlite3
import random
import time
import datetime
import matplotlib.pyplot as plt

#Esse comando abaixo indica que o grafico será exibid odentro desse notebook. Caso isso nãabso seja incluido
#O grafico será exibido em outra janela
%matplotlib notebook

conn = sqlite3.connect('dsa.db')

c = conn.cursor()

def dados_graficos():
    c.execute('SELECT id, valor FROM produtos')
    ids = []
    valores = []
    dados = c.fetchall()
    
    for l in dados:
        ids.append(l[0])
        valores.append(l[1])
    
    plt.bar(ids, valores)
    plt.show()
```

```python
dados_graficos()
```
